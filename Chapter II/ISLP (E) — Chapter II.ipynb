Use the `plot.hist()` method of college to produce some histograms with differing numbers of bins for a few of the quantitative variables.
```python
numeric_columns = df.select_dtypes(include=np.number).columns.tolist()
numeric_columns
```

```python
n = len(numeric_columns)

fig, axes = plt.subplots(n, 1, figsize=(14/2.54, 15))
fig.text(0.5, 0.04, 'Histogram of numerical columns', ha='center', va='center')

for ax, col in zip(axes, numeric_columns):
    sns.histplot(df.loc[:, col], ax=ax)
    ax.axvline(df.loc[:, col].mean(), c='k') 
    
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=2, wspace=0.4,hspace=0.8)
```
