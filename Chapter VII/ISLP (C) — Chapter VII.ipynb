### Exercise 7.9

It was mentioned in this chapter that a cubic regression spline with one knot at $\xi$ can be obtained using a basis of the form $x, x^{2}, x^{3}, (x - \xi)_{+}^3$, where $(x-\xi)^{3}_{+} = (x - \xi)^{3} \text{ if } x \gt \xi$ and equals $0$ otherwise. We will now show that a function of the form 

$$
f(x) = \beta_0 + \beta_1x + \beta_2x^{2} + \beta_3x^{2} + \beta_4(x - \xi)^{3}_{+}
$$

is indeed a cubic regression spline, regardless of the values of $\beta_0, \beta_1, \beta_2, \beta_3, \beta_4$.

#### Exercise 7.9.1 a

Find a cubic polynomial 

$$
f_1(x) = a_1 + b_1x + c_1x^{2} + d_1x^{3}
$$

such that $f(x) = f_1(x)$ for all $x \ge \xi$. Express $a_1, b_1, c_1, d_1$ in terms of $\beta_0, \beta_1, \beta_2, \beta_3, \beta_4$.

__Answer__:

Based on that polynomial $f_1(x)$ we can substitute our $\beta_0, \beta_1, \beta_2, \beta_3, \beta_4$:

$$
\begin{align*}
f(x) &= \beta_0 + \beta_1x + \beta_2x^{2} + \beta_3x^{3} + \beta_4(x - \xi)^{3} \\
&= \beta_0 + \beta_1x + \beta_2x^{2} + \beta_3x^{3} + 0 \text{ because } \forall x \le \xi \implies (x - \xi)^{3} = 0 \\
&= a_1 + b_1x + c_1x^{2} + d_1x^{3} \\
\end{align*}
$$

Let $a_1, b_1, c_1, d_1$ be defined as follows:

$$
a_1 = \beta_0, b_1 = \beta_1, c_1 = \beta_2, d_1 = \beta_3
$$

Then, we see:

$$
a_1 \land b_1 \land c_1 \land d_1 = \beta_3 \implies f(x) = f_1(x)
$$

### Exercise 7.9.1 b

Find a cubic polynomial:

$$
f_2(x) = a_2 + b_2x + c_2x^{2} + d_2x^{3}
$$

such that $f(x) = f_2(x)$ for all $x \gt \xi$. Express $a_2, b_2, c_2, d_2$ in terms of $\beta_0, \beta_1, \beta_2, \beta_3, \beta_4$. We have now established that $f(x)$ is a piecewise polynomial.

__Answer__:

Similarly as in the previous step, we substitute $\beta_0, \beta_1, \beta_2, \beta_3, \beta_4$:

$$
\begin{align*}
f(x) &= \beta_0 + \beta_1x + \beta_2x^{2} + \beta_3x^{3} + \beta_4(x - \xi)^{3} \text{ because } \forall x \gt \xi \implies (x-\xi)^{3} = (x-\xi)^{3} \\ 
&= \beta_0 + \beta_1x + \beta_2x^{2} + \beta_3x^{3} + \beta_4(-\xi^{3} + x^{3} - 3\xi x^{2} + 3\xi^{2}x) \\
&= (\beta_0 - \beta_4\xi^{3}) + (\beta_1 + \beta_43\xi^{2})x + (\beta_2 - \beta_43\xi)x^{2} + (\beta_3 + \beta_4)x^{3} \\
&= a_2 + b_2x + c_2x^{2} + d_2x^{3}
\end{align*}
$$

Let $a_2, b_2, c_2, d_2$ be defined as follows:

$$
a_2 = \beta_0 - \beta_4\xi^{3}, b_2 = \beta_1 + \beta_43\xi^{2}, c_2 = \beta_2 - \beta_43\xi, d_2 = \beta_3 + \beta_4
$$

Then, we got:

$$
a_2 \land b_2 \land c_2 \land d_2 \implies f(x) = f_2(x)
$$


#### Exercise 7.9.1 c

Show that $f_1(\xi) = f_2(\xi)$. That is, $f(x)$ is continuous at $\xi$. 

Let, first substitute for $f_1(x) = f_1(\xi)$ as we previousle have done it in the [7.9.1 (a)](#Exercise-7.9.1-a):

$$
f_1(\xi) = \beta_0 + \beta_1\xi + \beta_2\xi^{2} + \beta_3\xi^{3}
$$

Now, we also want to subsitute $x$ in piecewise polynomial by $\xi$:

$$
\begin{align*}
f_2(\xi) &= (\beta_0 - \beta_4\xi^{3}) + (\beta_1 + \beta_43\xi^{2})\xi + (\beta_2 - \beta_43\xi)\xi^{2} + (\beta_3 + \beta_4)\xi^{3} \\ 
&= \beta_0 - \beta_4\xi^{3} + \beta_1\xi + \beta_43\xi^{3} + \beta_2\xi^{2} - \beta_43\xi^{3} + \beta_3\xi^{3} + \beta_4\xi^{3} \\
&= \beta_0 + \beta_1\xi + \beta_2\xi^{2} + \beta_3\xi^{3}
\end{align*}
$$

Here, we can conclude that indeed $f_1(\xi) = f_2(\xi)$.

#### Exercise 7.9.1 d

Show that $f'_1(\xi) = f'_2(\xi)$. That is, $f'(x)$ is continuous at $\xi$.

__Answer__:

We first take the derivative of $f'_1(\xi)$:

$$
f'_1(\xi) = 0 + \beta_1 + 2\beta_2\xi + 3\beta_3\xi^{2} = \beta_1 + 2\beta_2\xi + 3\beta_3\xi^{2} 
$$

Then, we take first derivative of $f'_2(\xi)$:

$$
f'_2(\xi) = \beta_1 + 2\beta_2\xi + 3\beta_3\xi^{2}
$$

We therefore, conclude that $f'_1(\xi) = f'_2(\xi)$.

#### Exercise 7.9.1 e 

Show that $f''_1(\xi) = f''_2(\xi)$. That is, $f''(x)$ is continuous at $\xi$.

The second derivate of $f''_1(\xi)$ is:

$$
f''_1(\xi) = 2\beta_2 + 6\beta_3
$$

Therefore, the second derivative of $f''_2(\xi)$ must be equal to $f''_1(\xi)$.

So, in conclusion, we provide that $f(x)$ is continuous at $f'(x) \land f''(x)$. Therefore, we find that $f(x)$ is a cubic spline.

### Exercise 7.9.2

Suppose that a curve $\hat{g}$ is computed to smoothly fit a set of $n$ points using the following formula:

$$
\hat{g} = \arg \min_g \left( \sum_{i=1}^{n}(y_i - g(x_i))^{2} + \lambda \int \left[g^{(m)}(x)\right]^{2}dx \right)
$$

where $g^{(m)}$ represents the $m$th derivative of $g$ (and $g^{(0)} = g$). Provide example sketches of $\hat{g}$ in each of the following scenarios.

```python
import matplotlib.pyplot as plt 
import numpy as np
from scipy.interpolate import UnivariateSpline

# Generating now syntatic data
np.random.seed(0)
n = 100
x_data = np.sort(5 * np.random.rand(n))
y_data = 2 * np.cos(x_data) ** 2 - 3 * np.sin(x_data ** 2) + 0.1 * np.random.randn(n)  # Nonlinear function


plt.scatter(x_data, y_data, marker='x', color='magenta');
plt.plot(x_data, y_data, color='blue', label='$y$ true');
plt.legend();
```

#### Exercies 7.9.2 a

$\lambda = \infty, m = 0$.

```python
import matplotlib.pyplot as plt 
import numpy as np
from scipy.optimize import minimize

def loss_function(params, x, y, lambda_value, m):
    g_values = params[:len(x)]
    loss_data = np.sum((y - g_values)**2)  
    regularization_term = 0
    if lambda_value != 0:
        dx = g_values
        for _ in range(m): 
            dx = np.gradient(dx, x)
        regularization_term = lambda_value * np.trapz(dx ** 2, x)

    total_loss = loss_data + regularization_term
    return total_loss

def g(x):
    return 2 * np.cos(x) ** 2 - 3 * np.sin(x ** 2)

# Generating now syntatic data
np.random.seed(0)
n = 250
x_data = np.sort(5 * np.random.rand(n))
y_data = g(x_data) + 0.1 * np.random.randn(n)  # Nonlinear function

initial_params = np.ones(n)

lambda_value = np.inf
m_value = 0

result = minimize(loss_function, initial_params, args=(x_data, y_data, lambda_value, m_value), method='L-BFGS-B')
optimized_params = result.x

plt.scatter(x_data, y_data, label='Data points')
plt.plot(x_data, optimized_params[:n], label=f'Fitted curve (lambda={lambda_value}, m={m_value})', color='red')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.title('Fitting Curve with Custom Lambda and m')
plt.show();
```

### Exercise 7.9.3

Suppose we fit a curve with a basis functions $b_1(X) = X, b_2(X) = (X - 1)^{2}I(X \ge 1)$. (Note that $I(X \ge 1)$) equals to $1$ for $X \ge 1$ and $0$ otherwise). We fit the linear regression model 

$$
Y = \beta_0 + \beta_1b_1(X) + \beta_2b_2(X) + \epsilon
$$

and obtain coefficient estimates $\hat{\beta_0} = 1, \hat{\beta_1}=1,\hat{\beta_2}=-2$. Sketch the estimated curve between $X = -2$ and $X = 2$. Note the intercepts, slopes and other relevant information.

```python
def basis_function(X, term=1):
    if term == 1:
        return X
    elif term == 2:
        return np.where(X >= 1, np.square(X - 1), 0)
    
    return None

hat_beta_0 = 1
hat_beta_1 = 1 
hat_beta_2 = -2

epsilon = np.random.normal(size=100)
X = np.linspace(-2, 2, 100)
Y = hat_beta_0 + hat_beta_1 * basis_function(X) + hat_beta_2 * basis_function(X, term=2)

fig, ax = plt.subplots(figsize=(8,8))
plt.xlabel('$X \in [-2, 2]$')
plt.ylabel('Estimated $\hat{y}$')
plt.title('Estimated curve: $Y = \hat{\\beta}_0 + \hat{\\beta_1}\cdot b_1(X) + \hat{\\beta_2}\cdot b_2(X)$')
plt.grid('on')
plt.axvline(hat_beta_0, color='green', ls='--', label='$\hat{\\beta_0}$ - y-intercept')
plt.plot(X, Y, color='blue', linewidth=2, label='curve X vs. Y')
plt.legend();
```

### Exercise 7.9.4

Suppose we fit a cruve with a basis functions $b_1(X) = I(0 \le X \le 2) - (X-1)I(1 \le X \le 2), b_2(X) = (X - 3)I(3 \le X \le 4) + I (4 \lt X \le 5)$.

We fit a linear regression

$$
Y = \beta_0 + \beta_1b_1(X) + \beta_2b_2(X) + \epsilon
$$

and obtain coefficient estimates $\hat{\beta_0} = 1, \hat{\beta_1} = 1, \hat{\beta_2} = 3$. Sketch estimated curve between $X = -2$ and $X = 6$. Note the intercepts, slopes, and other releveant informations.

```python
def b1(X):
    return np.logical_and(0 <= X, X <= 2).astype(int) - (X - 1) * np.logical_and(1 <= X, X <= 2).astype(int)

def b2(X):
    return (X - 3) * np.logical_and(3 <= X, X <= 4).astype(int) + np.logical_and(4 < X, X <= 5).astype(int)

beta0_hat = 1
beta1_hat = 1
beta2_hat = 3

X_values = np.linspace(-2, 6, 1000)

Y_values = beta0_hat + beta1_hat * b1(X_values) + beta2_hat * b2(X_values)

plt.plot(X_values, Y_values, label='Estimated Curve')
plt.axhline(0, color='black', linestyle='--', linewidth=1)  

plt.text(0, beta0_hat, f' Intercept: {beta0_hat}', verticalalignment='bottom')
plt.text(1, beta0_hat + beta1_hat, f' Slope for $b_1$: {beta1_hat}', verticalalignment='bottom')
plt.text(4, beta0_hat + beta2_hat, f' Slope for $b_2$: {beta2_hat}', verticalalignment='bottom')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Estimated Curve from Linear Regression')
plt.legend()

plt.show()
```

### Exercise 7.9.5

Consider two curves, $\hat{g_1}$ and $\hat{g_2}$, defined by

$$
\hat{g_1} = \arg \min_g \left( \sum_{i=1}^{n}(y_i - g(x_i))^{2} + \lambda \int \left[g^{(3)}(x)\right]^{2} dx \right)
$$

$$
\hat{g_2} = \arg \min_g \left( \sum_{i=1}^{n}(y_i - g(x_i))^{2} + \lambda \int \left[g^{(4)}(x)\right]^{2} dx \right)
$$

where $g^{(m)}$ represents the $m$th derivative of $g$.

#### Exercise 7.9.5 a

As $\lambda \to \infty$, will $\hat{g_1}$ or $\hat{g_2}$ have the smaller training $\text{RSS}$?

__Answer__:

As $\lambda \to \infty$, both $\hat{g}_1$ and $\hat{g}_2$ will be heavily penalized for having non-zero derivatives. 
However, the penalty for $\hat{g}_2$ will be more severe because it involves the fourth derivative, which measures the rate of change of the curvature of the function.
As a result, $\hat{g}_2$ will be forced to be a polynomial of degree less than 4, which means it will be a cubic polynomial at most. 

On the other hand, $\hat{g}_1$ will be forced to be a polynomial of degree less than 3, which means it will be a quadratic polynomial at most. 

Since a cubic polynomial can fit more complex patterns in the data than a quadratic polynomial, $\hat{g}_2$ will likely have a smaller training $\text{RSS}$ than $\hat{g}_1$ as $\lambda \to \infty$.

#### Exercise 7.9.5 b

As $\lambda \to \infty$, will $\hat{g}_1$ or $\hat{g}_2$ have the smaller test $\text{RSS}$?

__Answer__:

As for the test $\text{RSS}$, it's not possible to say definitively without knowing more about the true relationship between the predictors and the response. If the true relationship is closer to a cubic polynomial, then $\hat{g}_2$ will likely have a smaller test $\text{RSS}$. If the true relationship is closer to a quadratic polynomial, then $\hat{g}_1$ will likely have a smaller test $\text{RSS}$. If the true relationship is very complex and non-linear, then both methods may have high test $\text{RSS}$ due to underfitting.

#### Exercise 7.9.5 c

For $\lambda = 0$, will $\hat{g}_1$ or $\hat{g}_2$ have the smaller training and test $\text{RSS}$?

__Answer__:

In both equations, the regularization term will be $0$, so it depends on the formula of $g$. The regularization term is $\lambda = 0$, so we can get a very wiggle or flexible function $\hat{g}$ which can produce very low training set, but very large test $\text{RSS}$ due to huge overfitting.

When we have the same knots, then we would have the same training $\text{RSS}$ for both functions $\hat{g}$, and the test $\text{RSS}$ would be very high for both $\hat{g}$ functions, due to large overfitting.
