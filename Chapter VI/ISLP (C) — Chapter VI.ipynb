### Exercise 6.6.5 

It is well-known that ridge regression tends to give similar coefficient values to correlated variables, whereas the lasso may give quite different coefficient values to correlated variables. We will now explore this property in a very simple setting.

Suppose that $n = 2$, $p = 2$, $x_{11} = x_{12}$, $x_{21} = x_{22}$. Furthermore, suppose that $y_1 + y_2 = 0$ and $x_{11} + x_{21} = 0$ and $x_{12} + x_{22} = 0$, so that the estimate for the intercept in a least squares, ridge regression, or lasso model is zero: $\hat{\beta} = 0$.

#### Exercise 6.6.5 a

Write out the ridge regression optimization problem in this setting.

__Answer__:

We know that Ridge Regression minimizes the training $\text{RSS}$:

$$
\underset{\beta}{\text{minimize }} \left( \sum_{i=1}^{n} \left(y_i - \beta_0 - \sum_{j=1}^{p}\beta_jx_{ij}\right)^{2} + \lambda\sum_{j=1}^{p}\beta_j^{2} \right)
$$

In our case, with the information that $\hat{\beta_0} = 0$, our Ridge Regression would minimize:

$$
\underset{\beta}{\text{minimize }} \left( (y_1 - (\hat{\beta_1}x_{11} + \hat{\beta_2}x_{12}))^{2} + (y_2 - (\hat{\beta_1}x_{21} + \hat{\beta_2}x_{22}))^{2} + \lambda \left(\hat{\beta_1}^{2} + \hat{\beta_2}^{2}\right) \right)
$$

Since we know that $x_{11} = x_{12}$ and $x_{21} = x_{22}$, the above equation can be simplified further to:

$$
\underset{\beta}{\text{minimize }} \left( (y_1 - x_{11}(\hat{\beta_1} + \hat{\beta_2}))^{2} + (y_2 - x_{21}(\hat{\beta_1} + \hat{\beta_2}))^{2} + \lambda \left(\hat{\beta_1}^{2} + \hat{\beta_2}^{2}\right) \right)
$$


#### Exercise 6.6.5 b

Argue that in this setting, the ridge coefficient estimates satisfy $\hat{\beta_1} = \hat{\beta_2}$.


__Answer__:

The loss function $\ell$ that we want to minimize is already known, i.e.:

$$
\ell(\hat{\beta_1}, \hat{\beta_2}) = \left( y_1 - x_{11} \cdot (\hat{\beta_1} + \hat{\beta_2}) \right)^{2} + \left( y_2 - x_{21} \cdot (\hat{\beta_1} + \hat{\beta_2}) \right)^{2} + \lambda \left(\hat{\beta_1}^{2} + \hat{\beta_2}^{2} \right)
$$

We should take the derivative of $\ell$ with respect to $\hat{\beta_1}$ and $\hat{\beta_2}$:

$$
\frac{\partial \ell}{\partial \hat{\beta_1}} = -2x_{11}\left(y_1 - x_{11}(\hat{\beta_1} + \hat{\beta_2})\right) - 2x_{21}\left(y_2 - x_{21}(\hat{\beta_1} + \hat{\beta_2}) \right) + 2\lambda\hat{\beta_1}
$$

$$
\frac{\partial \ell}{\partial \hat{\beta_2}} = -2x_{11}\left(y_1 - x_{11}(\hat{\beta_1} + \hat{\beta_2})\right) - 2x_{21}\left(y_2 - x_{21}(\hat{\beta_1} + \hat{\beta_2}) \right) + 2\lambda\hat{\beta_2}
$$

Setting both equation to 0, we got:

$$
\frac{\partial \ell}{\partial \hat{\beta_1}} = 0
$$

$$
\frac{\partial \ell}{\partial \hat{\beta_2}} = 0
$$

With given constraints, i.e.:
- $n = 2$,
- $p = 2$,
- $x_{11} = x_{12}$,
- $x_{21} = x_{22}$,
- $x_{11} + x_{21} = 0$,
- $x_{21} + x_{22} = 0$,
- $y_1 + y_2 = 0$;

We can make an assumption, that:

$$
X = \begin{bmatrix} x_{11} & x_{12} \\ x_{21} & x_{22} \end{bmatrix} = \begin{bmatrix} -1 & -1 \\ 1  &  1 \end{bmatrix}
$$

and 

$$y =  \begin{bmatrix} y_1 \\ y_2 \end{bmatrix} = \begin{bmatrix} 1 \\ -1 \end{bmatrix}$$

Now, we can plug into these values to our equation of partial derivatives:

$$
\begin{align*}
\frac{\partial \ell}{\partial \hat{\beta_1}} &= -2x_{11}\left(y_1 - x_{11}(\hat{\beta_1} + \hat{\beta_2})\right) - 2x_{21}\left(y_2 - x_{21}(\hat{\beta_1} + \hat{\beta_2}) \right) + 2\lambda\hat{\beta_1} \\
&= -2(-1) \left(1 - (-1)(\hat{\beta_1} + \hat{\beta_2})) \right) -2(1)\left(-1 -1(\hat{\beta_1} + \hat{\beta_2}) \right) + 2\lambda\hat{\beta_1} \\
&= 2 + 2\hat{\beta_1} + 2\hat{\beta_2} + 2 + 2\hat{\beta_1} + 2\hat{\beta_2} + 2\lambda\hat{\beta_1} \\
&= 4\hat{\beta_1} + 4\hat{\beta_2} + 4 + 2\lambda\hat{\beta_1} = 0
\end{align*}
$$

$$
\begin{align*}
\frac{\partial \ell}{\partial \hat{\beta_2}} &= -2x_{11}\left(y_1 - x_{11}(\hat{\beta_1} + \hat{\beta_2})\right) - 2x_{21}\left(y_2 - x_{21}(\hat{\beta_1} + \hat{\beta_2}) \right) + 2\lambda\hat{\beta_2} \\
&= -2(-1) \left(1 - (-1)(\hat{\beta_1} + \hat{\beta_2})) \right) -2(1)\left(-1 -1(\hat{\beta_1} + \hat{\beta_2}) \right) + 2\lambda\hat{\beta_2} \\
&= 2 + 2\hat{\beta_1} + 2\hat{\beta_2} + 2 + 2\hat{\beta_1} + 2\hat{\beta_2} + 2\lambda\hat{\beta_2} \\
&= 4\hat{\beta_1} + 4\hat{\beta_2} + 4 + 2\lambda\hat{\beta_2} = 0
\end{align*}
$$

We take this both equations, and setting them together:

$$
4\hat{\beta_1} + 4\hat{\beta_2} + 4 + 2\lambda\hat{\beta_1} = 4\hat{\beta_1} + 4\hat{\beta_2} + 4 + 2\lambda\hat{\beta_2}
$$

We can simplify it:

$$
\cancel{4(\hat{\beta_1} + \hat{\beta_2} + 1)} + 2\lambda\hat{\beta_1} = \cancel{4(\hat{\beta_1} + \hat{\beta_2} + 1)} + 2\lambda\hat{\beta_2} 
$$

We divide each side by $2\lambda$:

$$
\hat{\beta_1} = \hat{\beta_2}
$$

Q.E.D

#### Exercise 6.6.5 d

Argue that in this setting, the lasso coefficients $\hat{\beta_1}$ and $\hat{\beta_2}$ are not unique — in other words, there are many possible solutions to the optimization problem in [(c)](#Exercise-6.6.5-c). Describe these solutions.

__Answer__:

As we can see on the contour plot, there are many solutions for $\hat{\beta_1}$ and $\hat{\beta_2}$, since predictors are perfectly correlated ($x_{11} = x_{12}$ and $x_{21} = x_{22}$), there are many points on the edge of Diamond Area (blue line) that could potentially minimize the residual sum of squares.

In other words, any point along the edge of the diamond where the contours of the residual sum of squares are tangent to the diamond could be a potential solution. This is why there are many possible solutions for the lasso coefficients $\hat{\beta_1}$ and $\hat{\beta_2}$ in this setting.


```python
import numpy as np
import matplotlib.pyplot as plt

beta1 = np.linspace(-10, 10, 400)
beta2 = np.linspace(-10, 10, 400)

B1, B2 = np.meshgrid(beta1, beta2)

def objective(beta1, beta2, x11, x21, y1, y2):
    return (y1 - x11 * (beta1 + beta2))**2 + (y2 - x21 * (beta1 + beta2))**2

x11 = x12 = -1
x21 = x22 = 1
y1 = 1
y2 = -1

Z = objective(B1, B2, x11, x21, y1, y2)

s = 5
constraint = np.abs(B1) + np.abs(B2) - s

plt.figure(figsize=(8, 6))
contour = plt.contour(B1, B2, Z, levels=np.logspace(0, 5, 35), cmap='RdGy')
plt.clabel(contour, inline=1, fontsize=10)
plt.plot([0], [0], 'o', color='green', markersize=10)  
plt.contour(B1, B2, constraint, levels=[0], colors='b')
plt.title('Contour plot of the objective function and the constraint')
plt.xlabel('beta1')
plt.ylabel('beta2')
plt.grid(True)
plt.show()
```

#### Exercise 6.6.6 b

Consider (6.13) with $p = 1$. For some choice of $y_1$ and $\lambda \gt 0$, plot (6.13) as a function of $\beta_1$. Your plot should confirm that (6.13) is solved by (6.15).

```python
y1 = 2
beta1 = np.linspace(-10, 10, 400)
lambda_values = 3

def cost(beta1, y1, lambda_):
    return (y1 - beta1) ** 2 + lambda_ * abs(beta1)

def lasso_estim(y, lambda_):
    if y > lambda_ / 2:
        return y - lambda_ / 2
    if y < - lambda_ / 2:
        return y + lambda_ / 2
    if abs(y) <= lambda_ / 2:
        return 0

costs = cost(beta1, y1, lambda_values)
vectorize_lasso_estim = np.vectorize(lasso_estim)
beta1_hat = vectorize_lasso_estim(y1, lambda_values)

plt.figure(figsize=(8, 6))
plt.plot(beta1, costs, label='Cost function')
plt.plot(beta1_hat, cost(beta1_hat, y1, lambda_values), 'rx', markersize=10, label='Minima')
plt.xlabel('beta1')
plt.ylabel('Cost')
plt.title('Cost function for LassoRegression with p=1')
plt.legend()
plt.grid(True)
plt.show()
```
